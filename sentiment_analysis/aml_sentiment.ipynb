```python
########part c########
ylp_zero = []    
ylp_one = []    
for item in mylist1:
    if int(item[-1]) == 0:
        ylp_zero.append(item)
    else:
        ylp_one.append(item)
            
amz_zero = []    
amz_one = []    
for item in mylist2:
    if int(item[-1]) == 0:
        amz_zero.append(item)
    else:
        amz_one.append(item)
         
imbd_zero = []    
imbd_one = []    
for item in mylist3:
    if int(item[-1]) == 0:
        imbd_zero.append(item)
    else:
        imbd_one.append(item)
        
training = [0 for i in range(400)]*6
testing = [0 for i in range(100)]*6

for i in range(400):
    training[i] = ylp_zero[i]
    training[i + 400] = ylp_one[i]
    training[i + 800] = amz_zero[i]
    training[i + 1200] = amz_one[i]
    training[i + 1600] = imbd_zero[i]
    training[i + 2000] = imbd_one[i]
    
for i in range(100):
    testing[i] = ylp_zero[i + 400]
    testing[i + 100] = ylp_one[i + 400]
    testing[i + 200] = amz_zero[i + 400]
    testing[i + 300] = amz_one[i + 400]
    testing[i + 400] = imbd_zero[i + 400]
    testing[i + 500] = imbd_one[i + 400]
    
```

```python
########part d########
from nltk.stem import WordNetLemmatizer
from nltk import corpus
import string

wnl = WordNetLemmatizer()
def lemmatize(word):
    for wd_tp in ['a', 's', 'r', 'n', 'v']:
        lemmatize_word = wnl.lemmatize(word,wd_tp)
        if word != lemmatize_word:
            word = lemmatize_word
            break
    return word
```

```python
def dim_reduce(mtr, d):
    
    turn_mtr = np.array(mtr)
    my_lst = [] 
    
    for p in range(turn_mtr.shape[0]):
        lst = []
        for k in range(d):
            temp_x = np.array(turn_mtr[p,:])
            temp_v = np.array(VT[k,:])
            dot_pd = np.dot(temp_x,temp_v)
            lst.append(dot_pd)
        my_lst.append(lst) 
    f = np.array(my_lst, dtype=float)   
    
    return f
  
```
